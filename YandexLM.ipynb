```python
import sys
import pandas as pd
eat_places = pd.read_csv('datasets_eat_places.csv')
```

```python
eat_places.info()
```

```python
# Убираем заведения где средний чек неизвестен или больше 2500
eat_places = eat_places[(eat_places.average_bill.notna()) & (eat_places.average_bill <= 2500)]

# Заменяем все NaN на средние значения
eat_places.fillna(eat_places.mean(numeric_only=True))
```

```python
# Заведения из Петербурга
print(len(eat_places[eat_places.city == 'spb']))

# Заведения из Москвы
print(len(eat_places[eat_places.city == 'msk']))

# Также это можно было сделать так
# print(eat_places.city.value_counts())
```

```python
# Какие типы заведений есть в датасете
print(eat_places['rubric'].unique())
```

```python
# 'Бар, паб' ; 'Ресторан'

# Суммарное кол-во ресторанов и пабов
print(len(eat_places[(eat_places.rubric == 'Бар, паб') | (eat_places.rubric == 'Ресторан')]))

#Средний чек ресторанов
print(eat_places[eat_places.rubric == 'Ресторан']['average_bill'].mean())

#Средний чек пабов
print(eat_places[eat_places.rubric == 'Бар, паб']['average_bill'].mean())

#Средний чек ресторанов и пабов
print(eat_places[(eat_places.rubric == 'Бар, паб') | (eat_places.rubric == 'Ресторан')]['average_bill'].mean())
```

```python
# Какие заведения 'Кофе с собой' дейстыительно имеют эту фичу

print(len(eat_places[eat_places.rubric == 'Кофе с собой']) - len(eat_places[(eat_places.rubric == 'Кофе с собой') & (eat_places.coffee_to_go == 1)]))
```

```python
# Разница между средним чеком кафе в Москве и Санк-Петербурге

print(abs(eat_places[eat_places.city == 'msk']['average_bill'].mean() - eat_places[eat_places.city == 'spb']['average_bill'].mean()))
```

```python
eat_places.describe()

# Самая редкая фича - возможность приводить питомцев (0.075 %)
# Вторая по редкости (и это странно) - имение заведением кондиционера (0.093%)

# Картой можно оплатить почти везде(87%)
```

```python
# Новая таблица с кафе из Санкт-Петербурга
eat_places_spb = eat_places[eat_places.city == 'spb']
eat_places_spb
```

```python
# Средней чек ресторанов:
print(eat_places_spb['average_bill'].mean())

# Был: 750, стал: 689.5; Уменшился на 60,5
```

```python
eat_places_spb.describe()
```

```python
#Разница между средними значениями

all_avg = eat_places.describe().loc['mean']

spb_avg = eat_places_spb.describe().loc['mean']
print("значения: ")
print(pd.concat([all_avg.rename('all'), spb_avg.rename('spb')], axis=1))
print("Разница: ")
print(all_avg - spb_avg)

```
